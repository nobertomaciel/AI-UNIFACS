# **Algoritmo K-Means**
## **Documentação das bibliotecas utilizadas:**

https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

https://seaborn.pydata.org/

https://matplotlib.org/

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Carregar o dataset Iris
data = load_iris()
X = data.data
```

```python
# Aplicar K-Means com 3 clusters (porque sabemos que há 3 classes no Iris)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
# Criar um DataFrame para visualização
df = pd.DataFrame(X, columns=data.feature_names)
df['Cluster'] = labels
```

```python
# Plotar os clusters
labelsname = list()
for l in labels:
    labelsname.append(data.target_names[l]+"("+str(l)+")")

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labelsname, palette='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.legend()
plt.title("K-Means Clustering no Dataset Iris")
plt.show()
```
