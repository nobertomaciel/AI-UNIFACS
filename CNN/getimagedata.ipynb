```python
from PIL import Image
import requests
from io import BytesIO
```

```python
# url da imagem
url = "https://upload.wikimedia.org/wikipedia/commons/7/73/Lion_waiting_in_Namibia.jpg"

# baixa a imagem
response = requests.get(url)

# abre em binário
img = Image.open(BytesIO(response.content)).resize((28, 28))

# Usa getpixel() para acessar um pixel (ex: coordenada x=10, y=20)
x, y = 0, 0
pixel = img.getpixel((x, y))


print("Valores RGB do pixel em (",x,", ",y,"):", pixel)
print("R:", pixel[0])
print("G:", pixel[1])
print("B:", pixel[2])
```

```python
# pegando os pixels da imagem inteira
for x in range(img.width):
    for y in range(img.height):
        pixel = img.getpixel((x, y))
        print("Valores RGB do pixel em (",x,", ",y,"):", pixel)
```
