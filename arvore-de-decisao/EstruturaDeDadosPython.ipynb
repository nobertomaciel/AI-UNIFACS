```python
import pandas as pd
```

```python
# Listas (list):
# Ordenada
# Mutável
# Permite valores duplicados
# Acessa os elementos por índice
# Usada quando precisa de uma sequência ordenada e modificável de elementos.
numeros = [1, 2, 3, 4]
print("numeros[0]:",numeros[0])

# Dicionários (dict):
# Ordenado (a partir do Python 3.7)
# Mutável (pode ser alterado)
# Chaves únicas
# Usado quando precisa armazenar pares chave-valor para acesso rápido.
dados = {"nome": "Alice", "idade": 25, "notas": [420, 380, 390]}
print("dados[\"notas\"]:",dados["notas"])

# Tuplas (tuple):
# Ordenada
# Imutável (não pode ser alterada depois de criada)
# Permite valores duplicados
# Acessa os elementos por índice
# Usada quando precisa armazenar um conjunto fixo de valores que não serão modificados.
coordenadas = (10, 20)
print("coordenadas[0]:",coordenadas[0])

#Conjuntos (set):
# Não ordenado
# Mutável (pode adicionar/remover elementos)
# Não permite valores duplicados
# Não acessa os elementos por índice
# Usado quando precisa armazenar elementos únicos e realizar operações matemáticas como interseção e união.
valores = {1, 2, 3, 4}
valores.add(5)
print("valores:", valores)
```

```python
# DataFrames:
# Organizados em formato de tabela (linhas e colunas)
# Indexados automaticamente ou manualmente
# Podem conter diferentes tipos de dados em colunas distintas
# Permitem operações avançadas como filtragem, agregação e fusão de dados
# Baseados em arrays do NumPy, tornando-os eficientes para grandes volumes de dados
indices = ["day1", "day2", "day3"]
dados = {"calories": [420, 380, 390],"duration": [50, 40, 45]}
df = pd.DataFrame(dados,indices)

display(df)
```
